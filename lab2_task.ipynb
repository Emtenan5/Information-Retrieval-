```python
import nltk
from nltk.corpus import stopwords
nltk.download("stopwords")
stp=stopwords.words('english')
```

```python
import pandas as pd

pd.set_option('display.max_colwidth', 150)

import re

from snowballstemmer import stemmer

from tqdm import tqdm
```

```python
docs_df = pd.DataFrame([
    ["d0","This is the second lab of the information retrieval cycle"],
    ["d1","In this lab we will learn how to deal with English texts"],
    ["d2","Today is  29 September 2022 , Accompanying is Thursday 23 Rabi' al-Awal 1444"],
    ["d3","In this lab you will learn about the methods of preparing data and trimming English texts"]

],
           columns=["docID","raw_text"])
docs_df
```

```python
list(stp)
```

```python
len(stp)
```

```python
def remove_stopWords(sentence):

    terms=[]

    stopWords= set(stp)

    for term in sentence.split() : 

        if term not in stopWords :

           terms.append(term)

    return " ".join(terms)



docs_df["Removed_Stopword_text"]=docs_df["raw_text"].apply(remove_stopWords)

print("***************************************************************************documents after removing stopwords*********************************************************************")
docs_df
```

```python
def normalize(text):

    text = text.lower()
    text = re.sub("y", "i", text)
    text = re.sub("[,'-]", "", text)

    return(text)



docs_df["normalized_text"]=docs_df["Removed_Stopword_text"].apply(normalize)

print("***************************************************************************documents after normalizing*********************************************************************")

docs_df
```

```python
s = stemmer("english")

def stem(sentence):

    return " ".join([s.stemWord(x) for x in sentence.split()])

docs_df['Stemmer_text']=docs_df['normalized_text'].apply(stem)

print("***************************************************************************documents after stemming*********************************************************************")

docs_df
```
